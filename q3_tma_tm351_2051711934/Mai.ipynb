# *Visualizing the data set*

```python
data_frame['Happiness Score'].groupby(data_frame['Region']).describe()##Happiness score statistics based on region
```

```python
##Similar data that compares the rankings of countries within each region.
data_frame['Happiness Rank'].groupby(data_frame['Region']).describe()
```

```python
Happay = data_frame['Happiness Score'] ##define rate happeniess
economic = data_frame['Economy (GDP per Capita)']##define rate in column Economy (GDP per Capita)
```

```python
from scipy.stats import pearsonr    ## imports scipy to calculating the Pearson correlation coefficient between two variables.

correlation, var = pearsonr(Happay, economic)##calculating the Pearson correlation

print('Correlation efficient:', correlation)
print('p-value:', var)
```

```python
    ## define pivot_table
pivotable = pd.pivot_table(data_frame, values='Happiness Score', index='Region',
                             columns='Economy (GDP per Capita)', aggfunc='mean')

    ##define crostable
crostable = pd.crosstab(data_frame['Economy (GDP per Capita)'],pd.cut(data_frame['Happiness Score'], [1, 4, 6, 11]))

print('Pivot table:\n')##print all that
print(pivotable)

print('Crosstab:')
print(crostable)
```

```python
import seaborn as sae # create a scatter plot
sae.regplot(x=data_frame['Economy (GDP per Capita)'], y=data_frame['Happiness Score'])##A plot depicting a positive correlation,
                                                                        ##where the Happiness Score increases along with an increase in the Economy's GDP per Capita.
```

```python
import matplotlib.pyplot as p  ##creating a variety of plots and charts in Python import library.
##bar chart
joined = data_frame.groupby('Region')['Happiness Score'].mean()
p.bar(joined.index, joined.values)
p.xticks(rotation=90)
p.xlabel('Region')
p.ylabel('Mean Happiness Score')
p.title('Bar Chart of Mean Happiness Score by Region')
p.show()
```

```python
##create a box plot of the happiness score and economy
import matplotlib.pyplot as p  ##creating a variety of plots and charts in Python import library.
from scipy.stats import pearsonr    ## imports scipy to calculating the Pearson correlation coefficient between two variables.
sae.boxplot(x='Happiness Score', y='Economy (GDP per Capita)', data=data_frame)
p.xlabel('Happiness Score')
p.ylabel('Economy (GDP per Capita)')
p.title('Box Plot of Happiness Score and Economy')
p.show()
```

```python

from scipy.stats import pearsonr    ## imports scipy to calculating the Pearson correlation coefficient between two variables.
relation = data_frame.drop(['Happiness Rank', 'Country', 'Region'], axis=1).corr()
relation
##display correlation matrix
sae.heatmap(relation)
```

```python
import matplotlib.pyplot as p  ##creating a variety of plots and charts in Python import library.
from scipy.stats import pearsonr    ## imports scipy to calculating the Pearson correlation coefficient between two variables.
# built heat maps
pivot_table = data_frame.pivot_table(index='Region', columns='Economy (GDP per Capita)',values='Happiness Score', aggfunc='mean')
sae.heatmap(pivot_table, cmap='YlGnBu')
p.xlabel('Economy (GDP per Capita)')
p.ylabel('Region')
p.title('Heatmap of Happiness Score vs. Economy by Region')
p.show()
```
